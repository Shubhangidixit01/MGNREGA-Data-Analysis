## ADVANCED PREPROCESSING

### CREATING A SAFE WORKING COPY

```python
df_clean = df.copy()
```

### LOGICAL RANGE VALIDATION

```python
cols = ['percent_of_NRM_Expenditure',
        'percent_of_Expenditure_on_Agriculture_Allied_Works',
        'percentage_payments_generated_within_15_days']

for col in cols:
    df_clean.loc[(df_clean[col] < 0) | (df_clean[col] > 100), col] = np.nan
```

### OUTLIER DETECTION

```python
numeric_cols = df_clean.select_dtypes(include=['int64','float64']).columns

for col in numeric_cols:
    
    Q1 = df_clean[col].quantile(0.25)
    Q3 = df_clean[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    outliers = df_clean[(df_clean[col] < lower) | (df_clean[col] > upper)]
    
    print(f"{col} : {len(outliers)} outliers")
```

### TREATING LOGICALLY IMPOSSIBLE VALUES

#### 1. WAGE RATE

```python
df_clean.loc[df_clean['Average_Wage_rate_per_day_per_person'] < 0,
             'Average_Wage_rate_per_day_per_person'] = np.nan
#Wage cannot be negative. If found, the values will be converted to NAN
```

#### 2. EXPENDITURE VARIABLE

```python
exp_cols = ['Total_Exp','Total_Adm_Expenditure','Material_and_skilled_Wages','Wages']

for col in exp_cols:
    df_clean.loc[df_clean[col] < 0, col] = np.nan
#Expenditure values cannot be negative.
```

#### 3. PERSONDAYS AND WORKER COUNTS

```python
count_cols = [
'SC_persondays','ST_persondays','Women_Persondays',
'Total_Households_Worked',
'Total_Individuals_Worked','Total_No_of_Workers'
]

for col in count_cols:
    df_clean.loc[df_clean[col] < 0, col] = np.nan
#Counts cannot be negative.
```

### CREATING STATE-YEAR AGGREGATE DATASET

```python
df_clean.columns
```

```python
# State-year aggregation including operational performance indicators
df_state_year = df_clean.groupby(['state_name','fin_year'], as_index=False,observed=False).agg({
    
    'Total_Exp':'sum',
    'Approved_Labour_Budget':'sum',
    'Average_Wage_rate_per_day_per_person':'mean',
    'Total_Households_Worked':'sum',
    'Total_Individuals_Worked':'sum',
    'Average_days_of_employment_provided_per_Household':'mean',
    'Persondays_of_Central_Liability_so_far':'sum',
    'SC_persondays':'sum',
    'ST_persondays':'sum',
    'Women_Persondays':'sum',
    'Differently_abled_persons_worked':'sum',
    'Material_and_skilled_Wages':'sum',
    'Wages':'sum',
    'Total_Adm_Expenditure':'sum',
    'Total_No_of_HHs_completed_100_Days_of_Wage_Employment':'sum',
    'percentage_payments_generated_within_15_days':'mean',
    # Operational performance indicators
    'Number_of_Completed_Works':'sum',
    'Number_of_Ongoing_Works':'sum',
    'Total_No_of_JobCards_issued':'sum',
    'Total_No_of_Works_Takenup':'sum'
})
```

### CREATING REGION-YEAR AGGREGATED DATASET

```python
# Mapping states to regions
state_to_region = {
    'ANDHRA PRADESH': 'South',
    'RAJASTHAN': 'North',
    'BIHAR': 'East',
    'ASSAM': 'Northeast',
    'MAHARASHTRA': 'West',
    'MADHYA PRADESH': 'Central'
}

# Add region column
df_state_year['region'] = df_state_year['state_name'].map(state_to_region)
```

```python
# Region-year aggregation
df_region_year = df_state_year.groupby(['region','fin_year'], as_index=False,observed=False).agg({
    'Total_Exp':'sum',
    'Approved_Labour_Budget':'sum',
    'Average_Wage_rate_per_day_per_person':'mean',
    'Total_Households_Worked':'sum',
    'Total_Individuals_Worked':'sum',
    'Average_days_of_employment_provided_per_Household':'mean',
    'Persondays_of_Central_Liability_so_far':'sum',
    'SC_persondays':'sum',
    'ST_persondays':'sum',
    'Women_Persondays':'sum',
    'Differently_abled_persons_worked':'sum',
    'Material_and_skilled_Wages':'sum',
    'Wages':'sum',
    'Total_Adm_Expenditure':'sum',
    'Total_No_of_HHs_completed_100_Days_of_Wage_Employment':'sum',
    'percentage_payments_generated_within_15_days':'mean',
    # Operational performance indicators
    'Number_of_Completed_Works':'sum',
    'Number_of_Ongoing_Works':'sum',
    'Total_No_of_JobCards_issued':'sum',
    'Total_No_of_Works_Takenup':'sum'
})
```

### FEATURE ENGINEERING

#### 1. EXPENDITURE PER HOUSEHOLD

```python
df_region_year['exp_per_household'] = df_region_year['Total_Exp'] / df_region_year['Total_Households_Worked']
```

#### 2. WAGE SHARE RATIO

```python
df_region_year['wage_share_ratio'] = df_region_year['Wages'] / df_region_year['Total_Exp']
```

#### 3. SC/ST PARTICIPATION PERCENTAGE

```python
df_region_year['sc_participation_percent'] = df_region_year['SC_persondays'] / df_region_year['Persondays_of_Central_Liability_so_far'] * 100
df_region_year['st_participation_percent'] = df_region_year['ST_persondays'] / df_region_year['Persondays_of_Central_Liability_so_far'] * 100
```

#### 4. WOMEN PARTICIPATION RATIO

```python
df_region_year['women_participation_ratio'] = df_region_year['Women_Persondays'] / df_region_year['Persondays_of_Central_Liability_so_far'] * 100
```

#### 5. DIFFERENTLY-ABLED PARTICIPATION RATIO

```python
df_region_year['diff_abled_participation_ratio'] = df_region_year['Differently_abled_persons_worked'] / df_region_year['Total_Individuals_Worked'] * 100
```

#### 6. EMPLOYMENT FULFILLMENT RATIO

```python
df_region_year['employment_fulfillment_ratio'] = df_region_year['Average_days_of_employment_provided_per_Household'] / 100
```

#### 7. LABOUR BUDGET UTILIZATION RATIO

```python
df_region_year['labour_budget_utilization_ratio'] = df_region_year['Total_Exp'] / df_region_year['Approved_Labour_Budget']
```
