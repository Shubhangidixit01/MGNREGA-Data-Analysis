### OUTLIER DETECTION USING STATE-WISE IQR

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

#### WAGE RATE

```python
sns.boxplot(x=df['Average_Wage_rate_per_day_per_person'])
plt.title("Before Outlier Removal")
plt.show()
```

#### INSIGHTS

```python
def remove_outliers(group):
    
    Q1 = group['Average_Wage_rate_per_day_per_person'].quantile(0.25)
    Q3 = group['Average_Wage_rate_per_day_per_person'].quantile(0.75)
    
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 2.0 * IQR
    
    return group[(group['Average_Wage_rate_per_day_per_person'] >= lower) &
                 (group['Average_Wage_rate_per_day_per_person'] <= upper)]

df_state_iqr = df.groupby('state_name', group_keys=False, observed=False).apply(
    remove_outliers,
)
```

#### BOXPLOT OF WAGE RATE

```python
sns.boxplot(x=df_state_iqr['Average_Wage_rate_per_day_per_person'])
plt.title("After Outlier Removal")
plt.show()
```

#### INSIGHTS

```python
print("Original max:", df['Average_Wage_rate_per_day_per_person'].max())
print("Cleaned max:", df_state_iqr['Average_Wage_rate_per_day_per_person'].max())
```

```python
sns.boxplot(x=df_state_iqr['Approved_Labour_Budget'])
plt.title("Before Outlier Removal")
plt.show()
```

#### INSIGHT

```python
Q1 = df_state_iqr['Approved_Labour_Budget'].quantile(0.25)
Q3 = df_state_iqr['Approved_Labour_Budget'].quantile(0.75)

IQR = Q3 - Q1
upper = Q3 + 1.5 * IQR

df_state_iqr['Approved_Labour_Budget'] = df_state_iqr['Approved_Labour_Budget'].clip(upper=upper)
```

### FEATURE ENGINEERING

#### 1. EXPENDITURE PER HOUSEHOLD

```python
# Measures spending efficiency
df_state_year['exp_per_household'] = df_state_year['Total_Exp'] / df_state_year['Total_Households_Worked'].replace(0, np.nan)
```

#### 2. WAGE SHARE RATIO

```python
#Shows how much goes to labour vs material
df_state_year['wage_share_ratio'] = df_state_year['Wages'] / df_state_year['Total_Exp']
```

#### 3. SC/ST PARTICIPATION PERCENTAGE

```python
#Social inclusion indicator
df_state_year['sc_participation_percent'] = df_state_year['SC_persondays'] / df_state_year['Persondays_of_Central_Liability_so_far'] * 100
df_state_year['st_participation_percent'] = df_state_year['ST_persondays'] / df_state_year['Persondays_of_Central_Liability_so_far'] * 100
```

#### 4. WOMEN PARTICIPATION RATIO

```python
#Social inclusion metric
df_state_year['women_participation_ratio'] = df_state_year['Women_Persondays'] / df_state_year['Persondays_of_Central_Liability_so_far'] * 100
```

#### 5. DIFFERENTLY-ABLED PARTICIPATION RATIO

```python
#Inclusion metric
df_state_year['diff_abled_participation_ratio'] = df_state_year['Differently_abled_persons_worked'] / df_state_year['Total_Individuals_Worked'] * 100
```

#### 6. EMPLOYMENT FULFILLMENT RATIO

```python
#Measures how close to 100-day guarantee
df_state_year['employment_fulfillment_ratio'] = df_state_year['Average_days_of_employment_provided_per_Household'] / 100
```

#### 7. LABOUR BUDGET UTILIZATION RATIO

```python
#Efficiency indicator
df_state_year['labour_budget_utilization_ratio'] = df_state_year['Total_Exp'] / df_state_year['Approved_Labour_Budget']
```

#### 8. WORK COMPLETION EFFICIENCY 

```python
#Governance indicator
df_state_year['work_completion_ratio'] = df_state_year['Number_of_Completed_Works'] / df_state_year['Total_No_of_Works_Takenup']
```

```python
numeric_only = {
    'Total_Exp':'sum',
    'Approved_Labour_Budget':'sum',
    'Average_Wage_rate_per_day_per_person':'mean',
    'Total_Households_Worked':'sum',
    'Total_Individuals_Worked':'sum',
    'Average_days_of_employment_provided_per_Household':'mean',
    'Persondays_of_Central_Liability_so_far':'sum',
    'SC_persondays':'sum',
    'ST_persondays':'sum',
    'Women_Persondays':'sum',
    'Differently_abled_persons_worked':'sum',
    'Material_and_skilled_Wages':'sum',
    'Wages':'sum',
    'Total_Adm_Expenditure':'sum',
    'Total_No_of_HHs_completed_100_Days_of_Wage_Employment':'sum',
    'percentage_payments_generated_within_15_days':'mean',
    # Operational performance indicators
    'Number_of_Completed_Works':'sum',
    'Number_of_Ongoing_Works':'sum',
    'Total_No_of_JobCards_issued':'sum',
    'Total_No_of_Works_Takenup':'sum'
}

df_region_year = (
    df_state_year.groupby(['region','year'], observed=True)
    .mean(numeric_only=True)
    .reset_index()
)
```
