### CREATING A COPY OF DATASET FOR PREPROCESSING

```python
df = data.copy()
```

### CHECKING FOR NULL VALUES

```python
df.isna().sum()
```

### CHECKING FOR ZERO VALUES

```python
zero_counts = (df == 0).sum().sort_values(ascending=False)
zero_counts
```

### REPLACING ZERO VALUES WITH NAN

```python
df['Approved_Labour_Budget'] = df['Approved_Labour_Budget'].replace(0, np.nan)

df['Average_Wage_rate_per_day_per_person'] = df['Average_Wage_rate_per_day_per_person'].replace(0, np.nan)
```

```python
df
```

```python
mean_budget = df.groupby(['state_name','fin_year'])['Approved_Labour_Budget'].transform('mean')

df['Approved_Labour_Budget'] = df['Approved_Labour_Budget'].fillna(mean_budget)
```

```python
mean_wage = df.groupby(['state_name','fin_year'])['Average_Wage_rate_per_day_per_person'].transform('mean')

df['Average_Wage_rate_per_day_per_person'] = df['Average_Wage_rate_per_day_per_person'].fillna(mean_wage)
```

```python
(df['Approved_Labour_Budget'] == 0).sum()

(df['Average_Wage_rate_per_day_per_person'] == 0).sum()
```

### CONVERTING DATATYPES 

```python
df['fin_year'] = df['fin_year'].astype(str)
df['state_name'] = df['state_name'].astype('category')
df['district_name'] = df['district_name'].astype('category')
df['month'] = pd.Categorical(df['month'], 
                              categories=['Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec','Jan','Feb','Mar'],
                              ordered=True)
```
